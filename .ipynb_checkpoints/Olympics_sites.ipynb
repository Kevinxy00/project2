```python
# Set Dependencies
import pandas as pd	
```

```python
# url 
url_olympic_sites='https://en.wikipedia.org/wiki/List_of_Olympic_Games_host_cities'
```

```python
#Extract Olympic tables from wikipedia
df_olympic_sites_wiki=pd.read_html(url_olympic_sites)
```

```python
#olympics sites dataframe

df_olympic_sites_wiki= df_olympic_sites_wiki[0]
new_header=df_olympic_sites_wiki.iloc[0]
df_olympic_sites_wiki=df_olympic_sites_wiki[1:]
df_olympic_sites_wiki.columns=new_header
#df_olympic_sites_wiki.drop(df_olympic_sites_wiki.index[len(df_olympic_sites_wiki)-3])
df_olympic_sites_wiki=df_olympic_sites_wiki.iloc[43:62] # select specific rows
#df_olympic_sites_wiki=df_olympic_sites_wiki
df_olympic_sites_wiki
```

```python
# Export file as a CSV, without the Pandas index, but with the header
df_olympic_sites_wiki.to_csv("olympics_sites_wiki.csv", index=False, header=True)
```

```python
#Import CSV
olympic_sites = "olympics_sites.csv"
df_olympic_sites = pd.read_csv(olympic_sites, encoding = "ISO-8859-1")
df_olympic_sites['Location'] = "["+ df_olympic_sites['Latitude'].astype(str) + ", "+ df_olympic_sites['Longitude'].astype(str) + "]" 
df_olympic_sites.head()
```
